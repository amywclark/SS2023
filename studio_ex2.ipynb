```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
with open('heaney_beowulf.txt') as f:
    beowulf_text = f.read()
    beowulf_text_lower = beowulf_text.lower()
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
beowulf_tokenized = tokenizer.tokenize(beowulf_text_lower)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

print(stop_words)

beowulf_clean=[]
for w in beowulf_tokenized:
    if w not in stop_words:
        beowulf_clean.append(w)
```

```python
def ngrams(input_list, n):
    return zip(*[input_list[i:] for i in range(n)])
```

```python
from collections import Counter
Counter(list(ngrams(beowulf_clean, 3))).most_common(25)
```

```python
from collections import Counter
Counter(list(ngrams(beowulf_tokenized, 3))).most_common(25)
```

```python
import nltk.corpus  
from nltk.text import Text  
beo_conc = Text(beowulf_tokenized) 
beo_conc.concordance('son')
```
