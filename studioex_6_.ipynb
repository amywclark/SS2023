```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.text import Text
with open('arthur.txt') as f:
    arthur_read = f.read()
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
arthur_tokenized = tokenizer.tokenize(arthur_read)
arthur_text = Text(arthur_tokenized)
```

```python
import matplotlib.pyplot as plt
import numpy as np
arthur_text.dispersion_plot(["lady"])
```

```python
nltk.download('averaged_perceptron_tagger')
arthur_pos = nltk.pos_tag(arthur_tokenized)

pairs = []
word = [0]
tag = [1]
for (word1, tag1), (word2, tag2) in nltk.bigrams(arthur_pos):
    if word1.startswith("old") and tag2.startswith("NN"):
        pairs.append((word1, word2))

print(pairs)
```

```python
from collections import Counter
Counter(pairs).most_common(25)
```

```python
arthur_cs = Counter(pairs)

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame(arthur_cs.most_common()[:10])
df = df.set_index(0)
df.plot.barh()
plt.xlabel('Count')
plt.ylabel('Word')
plt.title('Word Frequency')
```

```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
             max_words=100,
        max_font_size=40, 
        scale=3,
        random_state=1 
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(arthur_tokenized)
```
